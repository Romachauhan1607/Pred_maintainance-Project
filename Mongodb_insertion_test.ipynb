```python
from pymongo import MongoClient
import pandas as pd

# MongoDB connection URI
uri = "mongodb+srv://miqbal:miqbal@cluster1.mr340pr.mongodb.net/?retryWrites=true&w=majority"

# Create a new client and connect to the server
client = MongoClient(uri)

# Database Name
db_name = "Nasa"
# Collection Names
collection_names = ["Train", "Test", "Rul"]

# Read CSV files
csv_file_paths = {
    "Train": "Notebook/RawData/Train1.csv",
    "Test": "Notebook/RawData/Test1.csv",
    "Rul": "Notebook/RawData/Rul1.csv",
}

# Insert data into MongoDB
try:
    # Iterate through the collection names and corresponding CSV files
    for collection_name, file_path in csv_file_paths.items():
        db = client[db_name]
        collection = db[collection_name]
        
        df = pd.read_csv(file_path)
        data = df.to_dict(orient='records')
        
        # Insert data in bulk
        collection.insert_many(data)
        
        print(f"Inserted {len(data)} documents into {collection_name} collection.")
    
    print("Data insertion completed successfully.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
finally:
    # Close the MongoDB client connection
    client.close()
```
